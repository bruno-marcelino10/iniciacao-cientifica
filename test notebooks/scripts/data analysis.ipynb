# Feature Import

```python
# Manipulação de Dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importação
dfs = pd.read_excel("../data/dados.xlsx", sheet_name = None, index_col = "Empresa")
print("DataFrames disponíveis: ", list(dfs.keys())) # representam indicadores de anos antes da falência
```

# Tratamento

```python
def importar_base(n):
    # Selecione dados de 1 a 5 anos pré-falência: 
    df = dfs[list(dfs.keys())[n-1]]
    print("\nDataFrame Escolhido:", n, "anos pré-falência")
    return df
```

# Análises

```python
def analise(df):
    desc = df.dropna().apply(pd.to_numeric).describe().T
    corr = df.dropna().apply(pd.to_numeric).corr()
    return desc, corr
```

# Escrevendo Planilhas

```python
analises = []
for i in range(1,6):
    df = importar_base(i)
    analises.append(analise(df))
```

```python
writer = pd.ExcelWriter('../data/stats.xlsx', engine='xlsxwriter')
for n in range(5):
    analises[n][0].to_excel(writer, sheet_name = list(dfs.keys())[n])
    
writer.save()
```

```python
writer = pd.ExcelWriter('../data/correl.xlsx', engine='xlsxwriter')
for n in range(5):
    analises[n][1].to_excel(writer, sheet_name = list(dfs.keys())[n])
    
writer.save()
```
